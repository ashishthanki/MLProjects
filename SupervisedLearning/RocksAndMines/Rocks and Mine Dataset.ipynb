# Split Data into Training and Testing Data

#### Use Stratified Sampling, but we need to turn 60th column into a category

If the value is 'R' (i.e. Rock) then replace it with zero, otherwise replace it with 1 (i.e. Mine) - note, Mine is not the True Positive. 

<i> You can use True Positive and True Negative interchangeably. As long as you know what your False Positive and False Negative is - which always confuses me!! - I always refresh myself on it before applying it in datasets. 

```python
import numpy as np
import random
from sklearn import datasets, linear_model
from sklearn.metrics import roc_curve, auc
import pylab as pl

df[60]=np.where(df[60] == 'R',0 ,1) 
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

def ss_data_frame(data):
    split = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=42)
    for train_index, test_index in split.split(df, df[60]):
        strat_train_set = df.loc[train_index]
        strat_test_set = df.loc[test_index]
    
    y_train = strat_train_set[60].copy()
    X_train = strat_train_set.drop(strat_train_set.columns[[60]], axis=1, inplace=False)
    y_test = strat_test_set[60].copy()
    X_test = strat_test_set.drop(strat_test_set.columns[[60]], axis=1, inplace=False)
    return(X_train, y_train, X_test, y_test)

X_train, y_train, X_test, y_test = ss_data_frame(df)
y_train.value_counts()/len(y_train)
```

# Standardisation

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer

number_pipeline = Pipeline([
    ('SS_Scaler', StandardScaler()),
])


number_attributes = X_train.columns

Full_pipeline = ColumnTransformer([
    ('Numbers', number_pipeline, number_attributes)
])

X_train_scaled = Full_pipeline.fit_transform(X_train)
X_test_scaled = Full_pipeline.transform(X_test)
```

# Train Classifier

##### Logical Regression seems like the best option

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(random_state=42)

log_model.fit(X_train, y_train)
```

```python
np.mean((log_model.predict(X_train) - y_train)**2)
```

```python
def precision_recall_confusion_details(X_train, y_train, y_predictions):
    print('Training Data R-Square: ', log_model.score(X_train, y_train))
    print('Test Data R-Square: ', log_model.score(X_test, y_test))

    print()

    from sklearn.metrics import precision_score, recall_score, f1_score
    print('Training Data Precision Score: ', precision_score(y_train, y_predictions))
    print('Training Data Recall Score: ', recall_score(y_train, y_predictions))

    print()

    print('Training Data F1 Score: ', f1_score(y_train, y_predictions))

    print()

    print('------------------Confusion Matrix------------------')
    from sklearn.metrics import confusion_matrix
    print(confusion_matrix(y_train, y_predictions))
    print()
    print('You can also obtain the values for TP, TN, FP and FN individually using the Sklearn\'s Confusion Matrix:')
    tn, fp, fn, tp = confusion_matrix(y_train, y_predictions).ravel()
    print('True Negative = '.rjust(20), tn)
    print('False Positive = '.rjust(20), fp)
    print('True Positive = '.rjust(20), tp)
    print('False Negative = '.rjust(20), fn)

    
    
y_predictions = log_model.predict(X_train)
precision_recall_confusion_details(X_train, y_train, y_predictions)
```

## Take a look at the results, what do they mean? 

Do we care more about precision or recall in the Rock and Mine content. In terms of Precision vs Recall, which one do we care more about (albeit, our results good). 

In this context, we know the following: 
<li> <b> True Postive: </b>means it predicts a Mine and it is a Mine - GOOD
<li> <b> True Negative:</b> means it predicts a Rock and it is a Rock - GOOD 
<li> <b> False Positive:</b> means it predicts a Mine but it is actually a Rock - Not ideal but we can live!
<li> <b> False Negative:</b> means it predicts a Rock but it is actually a Mine - # VERY BAD 

### What is your aim? 
    
If your aim is to <b> diffuse </b> Mines (and, obviously, not Rocks) then we may want to consider increasing precision further albeit, 83/5% is high enough, in my opinion, factoring in the good recall value we obtained.
    
However, if your aim is to <b> find all </b>the Mines knowing that a few Rocks will incorrectly be classified as Mines then you may want to increase Recall. 

#### Overall, the results are fairly good, but can we optimise the results through the threshold value? Logistical regression, by default, uses 0.5 but consider changing it? First analyse it using the cross validation predict function and return scores instead of predictions. This will help determine the fit of the line (i.e. overfitting or underfitting). 

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_recall_curve

y_train_pred = cross_val_predict(log_model, X_train, y_train, cv=3)
y_train_pred
precisions, recalls, thresholds = precision_recall_curve(y_train, y_train_pred)

precision_recall_confusion_details(X_train, y_train, y_train_pred)
```

### Cross Validation performed worse which suggests that the data is overfitting the Logistical Regression model, hence, why the result seem so good... 

```python
y_scores = cross_val_predict(log_model, X_train_scaled, y_train, cv=3, method= 'decision_function')
y_scores

precisions, recalls, thresholds = precision_recall_curve(y_train, y_scores)
```

```python
import matplotlib.pyplot as plt
#Plot a graph with the new scores 

def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
    plt.legend(loc="lower left", fontsize=16)
    plt.xlabel("Threshold", fontsize=16)        
    plt.grid(True)                              
    plt.axis([-2, 4, 0, 1.1])             

recall_90_precision = recalls[np.argmax(precisions >= 0.90)]
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]


plt.figure(figsize=(8, 4))                                                                  
plot_precision_recall_vs_threshold(precisions, recalls ,thresholds)
```

```python

def plot_precision_vs_recall(precisions, recalls):
    plt.plot(recalls, precisions, "b-", linewidth=2)
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.axis([0, 1.1, 0, 1.05])
    plt.grid(True)

plt.figure(figsize=(8, 6))
plot_precision_vs_recall(precisions, recalls)
plt.plot([0.9, 0.9], [0., 0.9], "r:")
plt.plot([0.0, 0.9], [0.9, 0.9], "r:")
plt.plot([0.9], [0.9], "ro")
plt.show()
```

#### We can improve this further, I think looking into a Support Vector Machine may be worthwhile. 

The logistical regression model does not seem stable and the Precision vs Recall graph is very far from smooth. This implies that any minor changes in the test (or new) data, will cause problems and massively affect Precision and Recall. 


```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV

svc = SVC(random_state=42, kernel= 'rbf')

params_distr = {'C': [6, 10, 30],
               'gamma': ['auto', 'scale']}

rnd_svc = RandomizedSearchCV(svc, random_state=42, param_distributions=params_distr, cv=5, n_jobs=-1)

rnd_svc.fit(X_train_scaled, y_train)
```

```python
from tensorflow import keras
import tensorflow as tf

model = keras.models.Sequential([
    keras.layers.Flatten(input_shape= [60]),
    keras.layers.Dense(300, activation= 'selu', kernel_initializer= 'lecun_normal'),
    keras.layers.Dense(100, activation= 'selu', kernel_initializer= 'lecun_normal'),
    keras.layers.Dense(1, activation= 'sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(loss= keras.losses.binary_crossentropy, optimizer= keras.optimizers.SGD(lr= 0.001),
              metrics= ['accuracy', keras.metrics.binary_accuracy])
model_history = model.fit(X_train_scaled, y_train, epochs=100,
                         validation_split=0.1,
                         callbacks= [keras.callbacks.EarlyStopping(patience=10)])
```

```python
pd.DataFrame(model_history.history).plot(figsize= (12, 6), title= 'Learning Curves')
```
